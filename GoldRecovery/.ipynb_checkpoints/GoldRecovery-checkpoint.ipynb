### Расчет эффективности обогащения

Формула для расчета эффективности обогощения:
$$
Recovery = \frac{C\bullet(F-T)}{F\bullet(C-T)}\bullet100 \%
$$

Нужные нам параметры по формуле:

- С - `rougher.output.concentrate_au` - доля золота после флотации
- F - `rougher.input.feed_au` - доля золота до флотации
- T - `rougher.output.tail_au` - доля золота в отвальных хвостах

```python
def recovery(C, F, T):
    return (C*(F-T))/(F*(C-T))*100
```

Рассчитаем эффктивность на тренировочной выборке:

```python
rated_recovery = recovery(
    gold_train['rougher.output.concentrate_au'], gold_train['rougher.input.feed_au'], gold_train['rougher.output.tail_au'])
```

Значение в столбце `rougher.output.recovery`:

```python
meaning_recovery = gold_train['rougher.output.recovery']
```

**MAE:**

```python
mean_absolute_error(meaning_recovery, rated_recovery)
```

МАЕ очень мала, поэтому значение эффективности в датасете можно счиать достоверным.

### Обработка пропусков

Для заполнения пропусков написана функция `next_fillna()`. Она нужна для того, чтобы заполнить пропуски в признаках соседними по времени параметрами в тренирвоочной и тестовой выборках.

<div style='color:gray;'>
Описание функции <b>next_fillna()()</b>: <br>
    <br>
    <b>Вход:</b>
        <ul>
            <li> data_sorted - отсортированные по дате данные </li>
        </ul>
    <b>Принцип работы:</b> <br>
   Функция проходит по каждому столбцу, в котором есть пропуски. Элементу,  соответствующему индексу пустого элемента призваивается среднее значение между соседними среднеми. Если какого-то соседнего значения нет или это пропуск, то заполнение происходит следующим или придыдущим элементом. 
        <br><br>
    <b>Выход:</b>
        <ol>
            <li> data_sorted - данные без пропусков </li>
        </ol>
</div>

```python
def next_fillna(data_sorted):
    columns = list(data_sorted.columns)
    for column in columns:
        null_index_list = data_sorted[data_sorted[column].isna()].index
        if len(null_index_list) > 0:
            size = data_sorted.shape[0]
            for i in null_index_list:
                if i == (size - 1) or pd.isnull(data_sorted.loc[i+1, column]):
                    mean_value = data_sorted.loc[i-1, column]
                elif i == 0 or pd.isnull(data_sorted.loc[i-1, column]):
                    mean_value = data_sorted.loc[i+1, column]
                else:
                    mean_value = (data_sorted.loc[i-1, column] + data_sorted.loc[i+1, column])/2
                data_sorted.loc[i, column] = mean_value
    return data_sorted
```

```python
gold_train_sorted = gold_train.sort_values(by='date')
gold_train = next_fillna(gold_train_sorted)
```

```python
gold_test_sorted = gold_test.sort_values(by='date')
gold_test = next_fillna(gold_test_sorted)
```

Рассчитаем метрику *sMAPE* по формуле:

$$
sMAPE=\frac{1}{N}\sum_\limits{i=1}^\limits{N}\frac{|y_i-\hat{y_i}|}{(|y_i|+|\hat{y_i}|)/2}\bullet100\% ,
$$
где $ y_i $ - реальное значение целевого признака, $ \hat{y_i} $ - предсказанное значение целевого признака <br>

Метрика выражается в процентах. Грубо говоря, метрика показывает наиболее точно, нежели MAPE, на сколько процентов в среднем предсказанное значение отличается от среднего. Поэтому наша цель - **минимальное значение метрики для модели.**

Функция для подсчета *sMAPE*:

```python
def smape(target, predictions):
    N = target.shape[0]
    numerator = abs(target - predictions) # подсчет числителя
    denominator = (abs(target) + abs(predictions))/2
    S = numerator/denominator
    return (S.sum()/N)*100
```

### Функции для кросс-валидации

Подготовим функции для кросс-валидации, в базовой функции `cross_val_score` есть возможность задать свою метрику св-вом `scoring`, т.к.метрика *sMAPE* не является базовой метрикой библиотеки `sklearn`, то зададим свой "скорер" с параметром `greater_is_better=False`:

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

<div style='color:gray;'>
Описание функции <b>cross_val_smape_lasso()</b>: <br>
    <br>
    <b>Вход:</b>
        <ul>
            <li>features - признаки </li>
            <li>target - целевой признак </li>
        </ul>
    <b>Принцип работы:</b> 
   Нахождение минимального sMAPE с разными размерами выборок в кросс-валидации для линейной регрессии. 
        <br><br>
    <b>Выход:</b>
        <ol>
            <li> минимальное значение sMAPE для линейной регрессии </li>
            <li> получившаяся наилучшая обученная модель линейной регрессии </li>
            <li> оптимальный cv </li>
        </ol>
</div>

```python
def cross_val_smape_lasso(features, target):
    cv_list = [3,5]
    score_smape_linear_min = 1_000_000_000
    for cv_ in cv_list:
        model_ = Lasso()
        smape_ = -cross_val_score(model_, features, target, cv=cv_, scoring=smape_scorer).mean()
        if smape_ < score_smape_linear_min:
            cv_linear = cv_
            model_linear = model_
            score_smape_linear_min = smape_
            
    return score_smape_linear_min, model_linear, cv_linear
```

<div style='color:gray;'>
Описание функции <b>cross_val_smape_forest()</b>: <br>
    <br>
    <b>Вход:</b>
        <ul>
            <li>features - признаки </li>
            <li>target - целевой признак </li>
            <li>depth_change - гиперпараметр: максимальная глубина дерева </li>
            <li>n_est_change - количество деревьев (число кратное 50)</li>
        </ul>
    <b>Принцип работы:</b> <br>
   Нахождение минимального sMAPE с разными размерами выборок в кросс-валидации для случайного леса. 
    <br><br>
    <b>Выход:</b>
        <ol>
            <li> минимальное значение sMAPE для случайного леса </li>
            <li> получившаяся наилучшая обученная модель случайного леса </li>
            <li> cv = 5 </li>
        </ol>
</div>

В силу долгого обучение случайного леса cv выбран равным 5, т.к. во всех других моделях это является оптимальным знаечнием.

```python
def cross_val_smape_forest(features, target, depth_change, n_est_change):
    score_smape_forest_min = 1_000_000_000
    cv_forest = 5
    for depth in range(2,depth_change):
        for n_est in range(50, n_est_change,50):
            model_ = RandomForestRegressor(random_state=RANDOM_STATE, max_depth=depth, n_estimators=n_est)
            smape_ = -cross_val_score(model_, features, target, cv=cv_forest, scoring=smape_scorer).mean()
            if smape_ < score_smape_forest_min:
                model_forest = model_
                score_smape_forest_min = smape_
                
    return score_smape_forest_min, model_forest, cv_forest
```

<div style='color:gray;'>
Описание функции <b>cross_val_smape_tree()</b>: <br>
    <br>
    <b>Вход:</b>
        <ul>
            <li>features - признаки </li>
            <li>target - целевой признак </li>
            <li>change_depth - глубина дерева </li>
        </ul>
    <b>Принцип работы:</b> <br>
   Нахождение минимального sMAPE с разными размерами выборок в кросс-валидации для случайного леса. <br><br>
    <b>Выход:</b>
        <ol>
            <li> среднее значение sMAPE наилучшего варианта валидационной и тренировочной выборок для решающего дерева </li>
            <li> минимальное значение sMAPE для решающего дерева </li>
            <li> получившаяся наилучшая обученная модель решающего дерева </li>
            <li> размер валидационной выборки </li>
        </ol>
</div>

```python
def cross_val_smape_tree(features, target, change_depth):
    score_smape_tree_min = 1_000_000_000
    for depth in range(2,change_depth):
        cv_list = [3,5]
        for cv_ in cv_list:
            model_ = DecisionTreeRegressor(random_state=RANDOM_STATE, max_depth=depth)
            smape_ = -cross_val_score(model_, features, target, cv=cv_, scoring=smape_scorer).mean()
            if smape_ < score_smape_tree_min:
                cv_tree = cv_
                model_tree = model_
                score_smape_tree_min = smape_
                
    return score_smape_tree_min, model_tree, cv_tree
```

Прежде чем делать окончательный вывод, необходимо посчитать итоговое sMAPE для флотации и очистки по формуле:<br>
<br>
$$
sMAPE=0.25 \bullet sMAPE(rougher) + 0.75 \bullet sMAPE(cleaner)

Для этого напишем функцию `smape_common()`:

### Проверка модели на тестовой выборке

Для оценки моделей напишем функцию `model_final()`.

<div style='color:gray;'>
Описание функции <b>model_final()</b>: <br>
    <br>
    <b>Вход:</b>
        <ul>
            <li>model_1 - первая модель </li>
            <li>model_2 - вторая модель</li>
            <li>features_1 - признаки первого набора данных </li>
            <li>target_1 - целевой признак первого набора данных </li>
            <li>features_2 - признаки второго набора данных </li>
            <li>target_2 - целевой признак второго набора данных </li>
        </ul>
    <b>Принцип работы:</b> <br>
   В функции моделируются предсказания каждой из моделей и высчитываются sMAPE каждой из них и общее sMAPE.
        <br><br>
    <b>Выход:</b>
        <ol>
            <li> sMAPE первой модели </li>
            <li> sMAPE второго модели </li>
            <li> общее sMAPE </li>
        </ol>
</div>

```python
def model_final(model_1, model_2, features_1, target_1, features_2, target_2):
    predictions_1 = model_1.predict(features_1)
    predictions_2 = model_2.predict(features_2)
    
    smape_1 = smape(target_1, predictions_1)
    smape_2 = smape(target_2, predictions_2)
    return smape_1, smape_2, smape_common(smape_1, smape_2)
```

Оценка на тестовой выборке:

```python
smape_rougher, smape_cleaner, smape_common_ = model_final(
    model_rougher, model_cleaner, features_test_rougher, target_test_rougher, features_test_cleaner, target_test_cleaner)
```

```python
print(f'sMAPE после этапа флотации : {smape_rougher:.4f}')
print(f'sMAPE после очистки : {smape_cleaner:.4f}')
print(f'Итоговое (общее) sMAPE : {smape_common_:.4f}')
```

```python
plt.bar(['Валидационная', 'Тестовая'], [smape_common_valid, smape_common_])
plt.title('Итоговое sMAPE в зависимости от выборок');
```

Процент, а который sMAPE тестовой отличается от валидационной:

```python
print(f'{abs(smape_common_valid - smape_common_):.2f} %')
```

В итоге мы получили, что значение sMAPE на валидационной выборке отличается на 1% от показателя на тестовой. 

Метрика показывает процентное значение разницы фактического и предсказанного значения, следовательно, средняя разница в **9%** является хорошим результатом (для полных выводов необходимо знать стандарты метрики для данной отрасли).

Проверим получкнную модель на адекватность с помощью константной модели:
